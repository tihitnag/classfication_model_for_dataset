```python
import math
import random
import numpy as np

data = []
with open('titanic/train.csv', 'r') as file:
    lines = file.readlines()
    lines = lines[1:]
    for line in lines:
        data.append(line.strip().split(","))

test = []
with open('titanic/test.csv', 'r') as file:
    lines = file.readlines()
    lines = lines[1:]
    for line in lines:
        test.append(line.strip().split(","))
 

def preprocess_data(data):
    for row in data:
        if row[6] == '':
            row[6] = '30'  # replace missing age with median age
        if row[12] == '':
            row[12] = 'S'
        row[5] = 0 if row[5] == 'male' else 1
        if row[12] == 'S':
            row[12] = 0
        elif row[12] == 'C':
            row[12] = 1
        elif row[12] == 'Q':
            row[12] = 2
        row[1] = float(row[1])
        row[2] = float(row[2])
        row[5] = int(row[5])
        row[6] = float(row[6])
        row[7] = int(row[7])
        row[8] = int(row[8])
        row[10] = float(row[10])
        
    return data



def preprocess_data_test(test):
    for row in test:
        if row[5] == '':
            row[5] = '30'  # replace missing age with median age
        if row[11] == '':
            row[11] = 'S'
        row[4] = 0 if row[4] == 'male' else 1
        if row[11] == 'S':
            row[11] = 0
        elif row[11] == 'C':
            row[11] = 1
        elif row[11] == 'Q':
            row[11] = 2
        row[1] = float(row[1])
        row[5] = float(row[5])
        row[6] = float(row[6])
        row[7] = int(row[7])
        
        if row[9] == '':
            row[9] = '0'
        row[9] = float(row[9])
        if row[5] == '':
            row[5] = '0'

        row[1] = float(row[1])
        row[5] = float(row[5])

        row[6] = int(row[6])
        
    return test


def sigmoid(z):
    return 1 / (1 + np.exp(-z))


# z=wx+b ...w-feature weight
def logistic_regression(feature, weight):
    z = 0
#     print("ooooooo", feature)
    for i in range(len(feature)):
#         feature[i] = float(feature[i])
#         print("kkkkkkkkk",type(feature[i]))
        
        if isinstance(feature[i], (int, float)):
            z += feature[i] * weight[i]
        else:
            raise TypeError(f"Expected feature[{i}] to be int or float, but got {type(feature[i])}")
   
    return sigmoid(z)


def loss( feature, weight):
    learning_rate = 0.01
    num_iterations = 10
#     print("hhhhhhhhhhhh")
    for _ in range(num_iterations):
        for j in range(len(weight)):
            weight[j] = random.uniform(-0.1, 0.1)
        gradient = [0] * len(weight)

        for row in data:
            feature = [row[1], row[2], row[5], row[6], row[7], row[8], row[10]]
            y_label = int(row[1])

            for j in range(len(weight)):
                gradient[j] += (logistic_regression(feature, weight) - y_label) * feature[j]

        for j in range(len(weight)):
            weight[j] = weight[j] - learning_rate * gradient[j]

        correct = 0

        for row in test:
#             print("yyyyyyyyyy", row[2])
            feature = [float(row[1]), int(row[4]), float(row[5]), int(row[6]), float(row[9])]

            y_label = int(row[1])
            prediction = round(logistic_regression(feature, weight))
            if prediction == y_label:
                correct += 1
        accuracy = correct / len(test)
        print("Accuracy:", accuracy)
        print('predtion :',prediction)
        print("correct",correct)

    return weight


data = preprocess_data(data)
test = preprocess_data_test(test)
loss([0]*7,[0]*7)
print(loss( [0] * 7, [0] * 7))
```
