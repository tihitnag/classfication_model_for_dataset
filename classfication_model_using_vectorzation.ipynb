```python
import numpy as np
import copy
import math
import numpy as np
import pandas as pd
import random
from sklearn.model_selection import train_test_split

# Load data
data=pd.read_csv('titanic.csv')

# Data preprocessing
x_data = data.drop(['Name', 'Ticket', 'Cabin'], axis=1)
x_data['Sex'] = x_data['Sex'].replace(['male', 'female'], [1, 0])
x_data['Embarked'].fillna('S', inplace=True)
x_data['Age'].fillna(30, inplace=True)
x_data['Embarked'] = x_data['Embarked'].replace(['S', 'C', 'Q'], [0, 1, 2])

# Split the data into training and testing sets
train_data, test_data = train_test_split(x_data, test_size=0.2, random_state=25)

# Convert data to numpy arrays
y_train = train_data['Survived'].values
x_train = train_data.drop(['Survived'], axis=1).values
y_test = test_data['Survived'].values
x_test = test_data.drop(['Survived'], axis=1).values






```

```python
def sigmoid(z):
    z = np.clip(z, -500, 500)
    g = 1 / (1 + np.exp(-z))
    return g
```

```python
def cost_function(x, y, w, b):
    m = x.shape[0]
    z = np.dot(x, w) + b
    f_wb = sigmoid(z)
    cost = -y * np.log(f_wb + 1e-10) - (1 - y) * np.log(1 - f_wb + 1e-10)
    return np.mean(cost)
```

```python
def gradient(x, y, w, b):
    m = x.shape[0]
    z = np.dot(x, w) + b
    f_wb = sigmoid(z)
    err = f_wb - y
    dw = np.dot(x.T, err) / m
    db = np.sum(err) / m
    return dw, db
```

```python
def logistic_regression(x_train, y_train, x_test):
    m_train, n_train = x_train.shape
    w = np.zeros(n_train)
    b = 0.
    alpha = 0.5
    iterations = 5000
    costs = []
    rand = random.uniform(-0.1, 0.1)
    w = np.full(n_train,0.0)
    for i in range(iterations):
        dw, db = gradient(x_train, y_train, w, b)
        w = w - alpha * dw
        b = b - alpha * db
        cost = cost_function(x_train, y_train, w, b)
        costs.append(cost)
        if i % 100 == 0:
            print('iteration %d cost: %f' % (i, cost))

    y_pred = []
    for i in range(x_test.shape[0]):
        z = np.dot(x_test[i], w) + b
        y_pred_i = sigmoid(z)
        if y_pred_i >= 0.5:
            y_pred.append(1)
        else:
            y_pred.append(0)

    # Calculate accuracy on test set
    accuracy = (y_pred == y_test).mean() * 100
    print(f"Accuracy: {accuracy:.2f}%")
    return y_pred, accuracy

y_pred, accuracy = logistic_regression(x_train, y_train, x_test)
print('Predictions:', y_pred)
print("Accuracy:", accuracy)
```
